```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()

data.target_names
```

```python
cat = ['talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=cat)
test = fetch_20newsgroups(subset='test', categories=cat)
```

```python
print(train.data[5])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(train.data, train.target)
labels = model.predict(test.data)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=train.target_names, yticklabels=train.target_names)
```

```python
def predictCatagory(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predictCatagory('sending a payload to the ISS')
```

```python
def cleanText(text):
    text = ''.join(char for char in text if char.isalnum() or char.isspace())    
    text = ' '.join(text.split()) # break the text into many other smaller parts
    return text

# the function called "isalnum" checks to see if a string is either a number
# or a string
#--------------------------------------------------------------------------#
# the function called "isspace" checks to see if there is a space or not in 
# the string
```

```python
# this line of code applies the cleanText function to the first index in the dataset
dataset[0] = dataset['v2'].apply(cleanText)
```
