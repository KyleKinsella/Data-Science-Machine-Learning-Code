```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="RdBu")
```

```python
print("value of X is:", X)
print("value of y is:", y)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X, y)
```

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="RdBu")
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap="RdBu", alpha=0.1)
```

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()

data.target_names
```

```python
cat = ['talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=cat)
test = fetch_20newsgroups(subset='test', categories=cat)
```

```python
print(train.data[5])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(train.data, train.target)
labels = model.predict(test.data)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=train.target_names, yticklabels=train.target_names)
```

```python
def predictCatagory(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predictCatagory('sending a payload to the ISS')
```
