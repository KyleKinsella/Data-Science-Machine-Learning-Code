```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("income.csv").dropna()
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
kmc = KMeans(n_clusters=3)
```

```python
kmc
```

```python
yPred = kmc.fit_predict(df[['Age', 'Income($)']])
```

```python
yPred
```

```python
df['cluster'] = yPred
```

```python
df.head()
```

```python
df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]
```

```python
plt.scatter(df0['Age'], df0['Income($)'], color='green')
plt.scatter(df1['Age'], df1['Income($)'], color='red')
plt.scatter(df2['Age'], df2['Income($)'], color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
## the scaler will make the graph be either 0 or 1. ##
scaler = MinMaxScaler()
```

```python
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
```

```python
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df
```

```python
kmc = KMeans(n_clusters=3)
yPred = kmc.fit_predict(df[['Age', 'Income($)']])
```

```python
yPred
```

```python
df['cluster'] = yPred
```

```python
df
```

```python
kmc.cluster_centers_
```

```python
df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]

plt.scatter(df0['Age'], df0['Income($)'], color='green')
plt.scatter(df1['Age'], df1['Income($)'], color='red')
plt.scatter(df2['Age'], df2['Income($)'], color='black')

plt.scatter(kmc.cluster_centers_[:, 0], kmc.cluster_centers_[:, 1], color='purple', marker='*', label='centroid')

plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
## elbow plot ##

k_rng = range(1, 10)
sse = []
for k in k_rng:
    kmc = KMeans(n_clusters=k)
    kmc.fit(df[['Age', 'Income($)']])
    sse.append(kmc.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse)
```
