```python
%matplotlib inline
import matplotlib.pyplot as plt
#print(plt.style.available)
plt.style.use("seaborn-v0_8-white")
import numpy as np 
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y) 
```

```python
# this is a test
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xFit = np.linspace(0, 10, 1000) 
yFit = model.predict(xFit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xFit, yFit)
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
Y = 0.5 + np.dot(X, [1.5, -2, 1.])

model.fit(X,Y)
print(model.intercept_)
print(model.coef_)
```

```python
from sklearn.preprocessing import PolynomialFeatures

x = np.array([2,3,4])

poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

```python
from sklearn.pipeline import make_pipeline

polyModel = make_pipeline(PolynomialFeatures(7), LinearRegression())
```

```python
import pandas as pd

file = "PremierLeagueSeason2024.csv"
read = pd.read_csv(file)

xVal = np.array(read["goals_scored"]).reshape(-1, 1)
yVal = read["goals_conceded"]


plt.xlabel("Goals Scored", size=15)
plt.ylabel("Goals Conceded", size=15)
plt.title("Goal's scored vs Goal' s Conceded")


from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

model.fit(xVal, yVal)

xFit = np.linspace(xVal.min(), xVal.max(), 1000).reshape(-1, 1)
yFit = model.predict(xFit)

plt.scatter(xVal, yVal)
plt.plot(xFit, yFit)
```
